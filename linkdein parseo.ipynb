```python
import pandas as pd
import random
from collections import Counter
```

```python
df = pd.read_csv('linke_scraper3.csv', index_col=0)
```

```python
exp = df.exp.to_list()
```

```python
len(exp)
```

```python
df.head(1)
```

# Clean positions

```python
all_positions = []
for experience in exp:
    
    f = experience

    f = f.split('\r\n')

    positions = []
    posicion = ['0']
    for e, text in enumerate(f):

        if f[e] != 'Nombre de la empresa':
            posicion.append(text)
        else:
            posicion.pop(-1)
            posicion = []
            positions.append(posicion)
            posicion.append(f[e-1])

    final_positions = []
    for pos in positions:
        if 'Duración total' in pos[2]:
            positions2 = []
            posicion2 = []
            company = pos[1]
            for e, text in enumerate(pos):
                if text != 'Fechas de empleo':
                    posicion2.append(text)
                else:
                    posicion2.pop(-1)
                    posicion2 = []
                    positions2.append(posicion2)
                    posicion2.append(pos[e-1])
                    posicion2.append(company)
                    posicion2.append('Fechas de empleo')
            final_positions += positions2
        else:
            final_positions.append(pos)

    all_positions.append(final_positions)
```

# Clean aptitudes

```python
apt1 = df.apt1.to_list()
```

```python
aptitudes = []
for apt in apt1:
    apti = []
    apt = apt.split('\r\n')
    if len(apt) == 1: 
        aptitudes.append(['nada'])
        continue
    if len(apt) > 1 and len(apt) < 4:
        aptitudes.append(apt)
        continue
    for e, text in enumerate(apt):
        if (('validado esta aptitud' in text) and (e != (len(apt) -1))):
            if ('aptitudes de LinkedIn' in apt[e+1]):
                if (e+1 != (len(apt) -1)):
                    apti.append(apt[e+2])
                else:
                    continue
            else:
                apti.append(apt[e+1])
        elif ('validaciones de' in text or 'validación de' in text or 'Aún no hay' in text):
            apti.append(apt[e-1])
    aptitudes.append(list(set(apti)))
```

```python
len(aptitudes)
```

# Aptitudes 2

```python
apt2 = df.apt2.to_list()
```

```python
aptitudes2 = []

remove = ['Conocimiento del sector', 'Herramientas y tecnologías', 'Otras aptitudes', 'Aptitudes interpersonales']

for apt in apt2:
    apt = apt.split('\r\n')
    apti2 = []
    for e, text in enumerate(apt):
        if ('validaciones' not in text) and ('validación' not in text):
            if text not in remove:
                try:
                    int(text)
                except:
                    apti2.append(text)
    aptitudes2.append(apti2)
```

# Info

```python
info = df['info'].to_list()
```

```python
name, location = [], []
for inf in info:
    inf = inf.split('\r\n')
    name.append(inf[0])
    location.append(inf[-1])
```

# New df

```python
ap = []
for e, i in zip(aptitudes, aptitudes2):
    ap.append(e + i)
```

```python
df_new = pd.DataFrame({'name':name, 'location':location, 'positions':all_positions, 'aptitudes':ap})
```
